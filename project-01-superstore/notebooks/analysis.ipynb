```python
import pandas as pd
import os

print(os.getcwd())

df = pd.read_csv('../data/Sample - Superstore.csv', encoding='latin-1')
# Q1 - Number of Rows and Cols
rows, cols = df.shape
print(f'Number of Rows: {rows}, Number of Columns: {cols}')
# Q2 - Data types of each column
print(df.dtypes)
# Q3 - Missing values in each column
print(df.isnull().values.any()) 
# Q4 - Category with highest Sales
print(df.groupby('Category')['Sales'].sum().sort_values(ascending=False))

# Q5 - Unique customers
print(df['Customer Name'].nunique())
# Q6 - Total Profit by Category
print(df.groupby('Category')['Profit'].sum().sort_values(ascending=False))

# Q7 - Profit Margin by Category (Profit divided by Sales)
df['Profit_Margin'] = df['Profit'] / df['Sales']
print(df.groupby('Category')['Profit_Margin'].mean().sort_values(ascending=False))

# Q8 - Which Sub-Category is LOSING money?
print("\n=== Q8 - Sub-Categories by Profit (Lowest First) ===")
print(df.groupby('Sub-Category')['Profit'].sum().sort_values())
```
