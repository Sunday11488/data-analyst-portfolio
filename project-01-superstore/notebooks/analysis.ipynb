```python
import pandas as pd
import os

print(os.getcwd())

df = pd.read_csv('../data/Sample - Superstore.csv', encoding='latin-1')
import matplotlib.pyplot as plt
import seaborn as sns

category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=category_profit.index, y=category_profit.values, palette='Blues_d')
plt.title('Total Profit by Category', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Total Profit ($)')
plt.tight_layout()
plt.savefig('../visuals/profit_by_category.png')
plt.show()
plt.savefig('../visuals/profit-category.png')
```

Profit by Category, To monitor and carefully observe the category profit relative to others


```python

plt.figure(figsize=(9,5))
plt.scatter(df['Discount'], df['Profit'], alpha=0.3, color='steelblue')
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.title('Does Discounting Kill Profit?', fontsize=14)
plt.xlabel('Discount Rate')
plt.ylabel('Profit ($)')
plt.tight_layout()
plt.savefig('../visuals/discount_vs_profit.png')
plt.show()
plt.savefig('../visuals/Discount-profit.png')
```

Scatter plot of discount per profit, It's exposes any loophole in it for fixing

```python
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values().head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=sub_profit.values, y=sub_profit.index, palette='Reds_r')
plt.title('Top 5 Loss-Making Sub-Categories', fontsize=14)
plt.xlabel('Total Profit ($)')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.savefig('../visuals/loss_making_subcategories.png')
plt.show()
plt.savefig('../visuals/toplosses.png')
```

# Superstore Sales Analysis

## Project Overview
A business analysis of 9,994 Superstore orders to identify 
profit drivers, loss-making products, and the impact of 
discounting on profitability.

## Business Questions Answered
- Which category drives the most profit - Technology
- Which sub-categories are losing money - Tables, Bookcases
- Does discounting hurt profitability - Yes, as it is seen in the visuals/ file

## Tools
Python · pandas · matplotlib · seaborn

## Key Findings
- Technology dominates profitability while Furniture lags significantly
- Tables, Bookcases, and Supplies generate negative profit; Tables are hardest hit at 26% average discount
- Discount rates show strong inverse correlation with profit—over-discounting is the primary profitability killer
- Reducing discounts on loss-making sub-categories is critical to restore Furniture category margins

Our analysis of 9,994 customer orders identifies a critical profitability challenge: Tables, Bookcases, and Supplies are all operating at a loss. Tables are the worst performer, with an average discount rate of 26%—the highest among all loss-making sub-categories. This excessive discounting is the primary driver of margin erosion. We recommend implementing a revised pricing strategy that limits discounts on these underperforming products. Without intervention, current discount levels will continue to erode Furniture category profitability.
